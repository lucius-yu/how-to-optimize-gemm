# Plot the performance curve for how to optimize gemm

```python
# import lib
import os
import glob
import pandas as pd
from pathlib import Path
```

```python
# load data
all_files = sorted(glob.glob("../data/*.csv"))
all_data=pd.DataFrame()

for file in all_files:
    data = pd.read_csv(file, usecols=['p', 'gflops'], index_col='p')
    data.columns=[Path(file).stem + '_gflops']
    all_data = pd.concat([all_data, data], axis=1)

display(all_data)
```

```python
all_data.plot(figsize=(10,8))
    
```
